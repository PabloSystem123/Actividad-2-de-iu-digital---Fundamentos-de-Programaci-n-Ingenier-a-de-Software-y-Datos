# (25%) Ejercicio 1
Vamos a mejorar la solución del Ejercicio 1 - Tarea 1.
Realice conversiones entre las unidades de medida presentadas en la siguiente tabla:

| Índice | Conversión                           | Factor |
|:------:|--------------------------------------|:------:|
| 0      | Kilómetro a metro                    | 1000   |
| 1      | Centímetro a metro                   | 0.01   |
| 2      | Kilómetro por hora a millas por hora | 0.62   |
| 3      | Nota en Educatic a nota en Canvas    | 20     |
| 4      | Nota en Canvas a nota en Educatic    | 0.05   |
| 5      | Mililitro a centímetro cúbico        | 1      |
| 6      | Dólar (USA) a peso colombiano        |Actualizar|
| 7      | Peso colombiano a dólar (USA)        |Actualizar|
| 8      | Libra a kilogramo                    | 0.45   |
| 9     | Minuto luz a kilómetro               | 17987539.67|

Para este ejercicio se deben almacenar los factores de conversión en un vector (tipo de dato ```list``` de Python) para luego:

- Solicitar al usuario que escoja el tipo de conversión que quiere realizar, para esto hay dos opciones:
 - Si trabajas en Google Colab puedes usar una lista desplegable, que es un tipo de campo de [formulario](https://www.youtube.com/watch?v=vx7yOZ5_Cw8) llamado *dropdown*
 - Si trabajas en otro editor puedes imprimir un menú de opciones numeradas y solicitarle al usuario que ingrese con ```input```el número de la opción que le interesa.

- Solicitar al usuario que ingrese el dato que quiere convertir.
- **Realizar una estructura de selección que convierta el dato ingresado por el usuario según el tipo de conversión que ha escogido.**

Al igual que en la Tarea 1, se debe acceder al factor de conversión correspondiente de acuerdo a su posición o índice en el vector de factores de conversión.


````python
# Desarrolle su ejercicio aquí

from IPython.display import display, HTML
from ipywidgets import widgets
import warnings
warnings.filterwarnings('ignore')

# Lista de factores de conversión
factores_conversion = [1000,    # 0: Kilómetro a metro
                      0.01,     # 1: Centímetro a metro
                      0.62,     # 2: Kilómetro por hora a millas por hora
                      20,       # 3: Nota en Educatic a nota en Canvas
                      0.05,     # 4: Nota en Canvas a nota en Educatic
                      1,        # 5: Mililitro a centímetro cúbico
                      4000,     # 6: Dólar a peso colombiano
                      0.00025,  # 7: Peso colombiano a dólar
                      0.45,     # 8: Libra a kilogramo
                      17987539.67] # 9: Minuto luz a kilómetro

# opciones de conversión
opciones_conversion = [
    "Kilómetro a metro",
    "Centímetro a metro",
    "Kilómetro por hora a millas por hora",
    "Nota en Educatic a nota en Canvas",
    "Nota en Canvas a nota en Educatic",
    "Mililitro a centímetro cúbico",
    "Dólar (USA) a peso colombiano",
    "Peso colombiano a dólar (USA)",
    "Libra a kilogramo",
    "Minuto luz a kilómetro"
]

def realizar_conversion(valor, indice):
    """
    Realiza la conversión según el índice seleccionado
    """
    try:
        valor = float(valor)
        resultado = valor * factores_conversion[indice]
        return resultado
    except ValueError:
        return "Error: Por favor ingrese un valor numérico válido"
    except IndexError:
        return "Error: Índice de conversión no válido"

def version_colab():
    """
    Versión con dropdown para Google Colab
    """
    dropdown = widgets.Dropdown(
        options=[(opcion, i) for i, opcion in enumerate(opciones_conversion)],
        description='Conversión:',
        style={'description_width': 'initial'},
        layout={'width': 'auto'}
    )

    entrada = widgets.FloatText(
        description='Valor:',
        style={'description_width': 'initial'}
    )

    salida = widgets.Output()

    def on_value_change(change):
        with salida:
            salida.clear_output()
            if entrada.value is not None:
                resultado = realizar_conversion(entrada.value, dropdown.value)
                print(f"Resultado: {resultado}")

    entrada.observe(on_value_change, names='value')
    dropdown.observe(on_value_change, names='value')

    display(dropdown)
    display(entrada)
    display(salida)

def version_consola():
    """
    Versión con menú para consola
    """
    while True:
        print("\nOpciones de conversión disponibles:")
        for i, opcion in enumerate(opciones_conversion):
            print(f"{i}. {opcion}")

        try:
            indice = int(input("\nIngrese el número de la conversión que desea realizar (0-9): "))
            if indice < 0 or indice >= len(opciones_conversion):
                print("Error: Por favor seleccione una opción válida (0-9)")
                continue

            valor = input(f"\nIngrese el valor a convertir de {opciones_conversion[indice]}: ")
            resultado = realizar_conversion(valor, indice)
            print(f"\nResultado: {resultado}")

            continuar = input("\n¿Desea realizar otra conversión? (s/n): ").lower()
            if continuar != 's':
                print("¡Gracias por usar el convertidor!")
                break

        except ValueError:
            print("Error: Por favor ingrese un número válido")
        except Exception as e:
            print(f"Error inesperado: {e}")

try:
    import google.colab
    in_colab = True
except:
    in_colab = False

if in_colab:
    print("Detectado entorno Colab. Usando interfaz con dropdown.\n")
    version_colab()
else:
    print("Ejecutando versión de consola.\n")
    version_consola()
````

# (25%) Ejercicio 2
Estudie cómo realizar conversiones entre los sistemas de numeración Hexadecimal, Binario y Decimal.

Programe un algoritmo que solicite al usuario un número entre 0 y 15 y entregue el valor correspondiente en los sistemas numéricos Hexadecimal y Binario.

**Nota:** No está permitido usar funciones de conversión nativas como ``bin()``, ``hex()``, etc, o funciones similares de librerías.

````python
# Desarrolle su ejercicio aquí

def decimal_a_binario(decimal):
    """
    Convierte un número decimal a binario usando división sucesiva.
    """
    if decimal == 0:
        return "0000"

    binario = ""
    numero = decimal

    while numero > 0:
        residuo = numero % 2
        binario = str(residuo) + binario
        numero = numero // 2

    while len(binario) < 4:
        binario = "0" + binario

    return binario

def decimal_a_hexadecimal(decimal):
    """
    Convierte un número decimal a hexadecimal usando división sucesiva.
    """
    if decimal == 0:
        return "0"

    hex_map = {
        10: 'A',
        11: 'B',
        12: 'C',
        13: 'D',
        14: 'E',
        15: 'F'
    }

    hexadecimal = ""
    numero = decimal

    while numero > 0:
        residuo = numero % 16
        if residuo > 9:
            hexadecimal = hex_map[residuo] + hexadecimal
        else:
            hexadecimal = str(residuo) + hexadecimal
        numero = numero // 16

    return hexadecimal

def validar_entrada(numero):
    """
    Valida que la entrada sea un número entre 0 y 15.
    """
    try:
        num = int(numero)
        if 0 <= num <= 15:
            return True, num
        else:
            return False, "El número debe estar entre 0 y 15"
    except ValueError:
        return False, "Por favor ingrese un número válido"

def main():
    print("   Conversor de Sistemas Numéricos  ")
    print()
    print("Convertidor de números decimales (0-15) a binario y hexadecimal")

    while True:
        entrada = input("\nIngrese un número entre 0 y 15 (o 'x' para salir): ")

        if entrada.lower() == 'x':
            print("\n¡Gracias por usar el conversor!")
            break

        valido, resultado = validar_entrada(entrada)

        if not valido:
            print(f"Error: {resultado}")
            continue

        decimal = resultado
        binario = decimal_a_binario(decimal)
        hexadecimal = decimal_a_hexadecimal(decimal)

        print(f"\nResultados para el número {decimal}:")
        print(f"Binario: {binario}")
        print(f"Hexadecimal: {hexadecimal}")
        print("-" * 40)

if __name__ == "__main__":
    main()
````

# (25%) Ejercicio 3

Propiedades de los triángulos.
*   Si tenemos tres segmentos de cualquier longitud, no siempre se puede construir un triángulo con ellos. En cualquier triángulo, la suma de dos de sus lados siempre es mayor que el tercer lado.
*   La suma de los ángulos de un triángulo es igual a 180°.

De acuerdo con la información descrita previamente, y haciendo uso de la función `input` realizar un algoritmo que reciba tres lados y tres ángulos y determine si puede dibujarse un triángulo, en caso contrario indicar qué  problemas presenta.



````python
# Desarrolle su ejercicio aquí

print("Verificador de triangulos")

def validar_entrada_numerica(mensaje):
    """
    Valida que la entrada sea un número positivo.
    """
    while True:
        try:
            valor = float(input(mensaje))
            if valor <= 0:
                print("Error: El valor debe ser mayor que 0")
                continue
            return valor
        except ValueError:
            print("Error: Por favor ingrese un número válido")

def validar_lados(lado1, lado2, lado3):
    """
    Verifica la desigualdad triangular para los tres lados.
    Retorna (bool, list): True si es válido, False si no, junto con lista de problemas.
    """
    problemas = []

    # Verificar las tres combinaciones de la desigualdad triangular
    if lado1 + lado2 <= lado3:
        problemas.append(f"La suma de los lados {lado1:.2f} y {lado2:.2f} debe ser mayor que {lado3:.2f}")
    if lado2 + lado3 <= lado1:
        problemas.append(f"La suma de los lados {lado2:.2f} y {lado3:.2f} debe ser mayor que {lado1:.2f}")
    if lado1 + lado3 <= lado2:
        problemas.append(f"La suma de los lados {lado1:.2f} y {lado3:.2f} debe ser mayor que {lado2:.2f}")

    return len(problemas) == 0, problemas

def validar_angulos(angulo1, angulo2, angulo3):
    """
    Verifica que los ángulos sumen 180° y sean positivos.
    Retorna (bool, list): True si es válido, False si no, junto con lista de problemas.
    """
    problemas = []

    # Verificar que cada ángulo sea menor que 180°
    if angulo1 >= 180:
        problemas.append(f"El ángulo {angulo1:.2f}° debe ser menor que 180°")
    if angulo2 >= 180:
        problemas.append(f"El ángulo {angulo2:.2f}° debe ser menor que 180°")
    if angulo3 >= 180:
        problemas.append(f"El ángulo {angulo3:.2f}° debe ser menor que 180°")

    # Verificar la suma de los ángulos
    suma = angulo1 + angulo2 + angulo3
    if abs(suma - 180) > 0.01:  # Usando una pequeña tolerancia para errores de punto flotante
        problemas.append(f"La suma de los ángulos debe ser 180°, pero es {suma:.2f}°")

    return len(problemas) == 0, problemas

def main():
    print("Por favor, ingrese las medidas del triángulo")

    # Solicitar los lados
    print("\nIngrese los lados del triángulo:")
    lado1 = validar_entrada_numerica("Lado 1: ")
    lado2 = validar_entrada_numerica("Lado 2: ")
    lado3 = validar_entrada_numerica("Lado 3: ")

    # Solicitar los ángulos
    print("\nIngrese los ángulos del triángulo (en grados):")
    angulo1 = validar_entrada_numerica("Ángulo 1: ")
    angulo2 = validar_entrada_numerica("Ángulo 2: ")
    angulo3 = validar_entrada_numerica("Ángulo 3: ")

    # Validar las propiedades
    lados_validos, problemas_lados = validar_lados(lado1, lado2, lado3)
    angulos_validos, problemas_angulos = validar_angulos(angulo1, angulo2, angulo3)

    # Mostrar resultados
    print("\nResultados del análisis:")
    print("-" * 50)

    if lados_validos and angulos_validos:
        print("¡El triángulo es válido! Cumple con todas las propiedades.")
        print(f"\nMedidas del triángulo:")
        print(f"Lados: {lado1:.2f}, {lado2:.2f}, {lado3:.2f}")
        print(f"Ángulos: {angulo1:.2f}°, {angulo2:.2f}°, {angulo3:.2f}°")
    else:
        print("El triángulo NO es válido. Se encontraron los siguientes problemas:")
        if not lados_validos:
            print("\nProblemas con los lados:")
            for problema in problemas_lados:
                print(f"- {problema}")
        if not angulos_validos:
            print("\nProblemas con los ángulos:")
            for problema in problemas_angulos:
                print(f"- {problema}")

if __name__ == "__main__":
    main()
````

# (25%) Ejercicio 4

Al final del bloque completaremos un programa que permita calcular la calificación final de un curso de acuerdo al porcentaje de ponderación y a la calificación obtenida en cada tarea.

En este ejercicio debes solicitar al usuario, con la función ```input``` o por medio de formularios, las ponderaciones de cada una de las 4 tareas de un curso. También, debes solicitarle la calificación obtenida y el nombre de cada tarea.

Toda la información adquirida debe ser almacenada en un diccionario (revisa los cuadernos de práctica y los módulos de Canvas donde se presenta el concepto). Luego, calcularás la calificación global del curso para finalmente presentar un resumen del desempeño del usuario.

````python
# Desarrolle su ejercicio aquí

from IPython.display import display, HTML
from ipywidgets import widgets
import warnings
warnings.filterwarnings('ignore')

def validar_porcentaje(valor, total_actual):
    """
    Valida que el porcentaje sea un número entre 0 y 100 y que la suma no exceda 100
    """
    try:
        porcentaje = float(valor)
        if porcentaje <= 0:
            return False, "El porcentaje debe ser mayor que 0"
        if porcentaje > 100:
            return False, "El porcentaje no puede ser mayor que 100"
        if total_actual + porcentaje > 100:
            return False, f"La suma de porcentajes excedería 100% (actual: {total_actual}%)"
        return True, porcentaje
    except ValueError:
        return False, "Por favor ingrese un número válido"

def validar_calificacion(valor):
    """
    Valida que la calificación sea un número entre 0 y 5
    """
    try:
        calificacion = float(valor)
        if calificacion < 0 or calificacion > 5:
            return False, "La calificación debe estar entre 0 y 5"
        return True, calificacion
    except ValueError:
        return False, "Por favor ingrese un número válido"

def calcular_nota_final(tareas):
    """
    Calcula la nota final ponderada
    """
    nota_final = 0
    for tarea in tareas.values():
        ponderacion = tarea['ponderacion'] / 100
        nota_final += tarea['calificacion'] * ponderacion
    return nota_final

def generar_reporte(tareas, nota_final):
    """
    Genera un reporte detallado del desempeño
    """
    reporte = "\n=== Reporte de Desempeño ===\n\n"
    reporte += "Desglose por tarea:\n"
    reporte += "-" * 50 + "\n"

    # Encontrar la calificación más alta y más baja
    max_nota = max(tareas.values(), key=lambda x: x['calificacion'])
    min_nota = min(tareas.values(), key=lambda x: x['calificacion'])

    for nombre, datos in tareas.items():
        reporte += f"Tarea: {nombre}\n"
        reporte += f"  Calificación: {datos['calificacion']:.2f}/5.00\n"
        reporte += f"  Ponderación: {datos['ponderacion']}%\n"
        reporte += f"  Aporte a nota final: {(datos['calificacion'] * datos['ponderacion']/100):.2f}\n"

        # Agregar indicadores de mejor/peor desempeño
        if datos == max_nota:
            reporte += "  ⭐ Mejor desempeño\n"
        if datos == min_nota:
            reporte += "  ⚠️ Oportunidad de mejora\n"
        reporte += "-" * 50 + "\n"

    # Agregar nota final y retroalimentación
    reporte += f"\nNota Final: {nota_final:.2f}/5.00\n"
    if nota_final >= 4.5:
        reporte += "¡Excelente desempeño! 🏆\n"
    elif nota_final >= 4.0:
        reporte += "Muy buen trabajo 👏\n"
    elif nota_final >= 3.0:
        reporte += "Has aprobado el curso ✅\n"
    else:
        reporte += "Necesitas mejorar para aprobar el curso ⚠️\n"

    return reporte

def version_colab():
    """
    Versión con widgets para Google Colab
    """
    tareas = {}
    total_ponderacion = 0

    # Crear widgets para cada tarea
    def crear_widgets_tarea(num_tarea):
        nombre = widgets.Text(description=f'Nombre Tarea {num_tarea}:', style={'description_width': 'initial'})
        ponderacion = widgets.FloatText(description='Ponderación (%):', style={'description_width': 'initial'})
        calificacion = widgets.FloatText(description='Calificación:', style={'description_width': 'initial'})
        return nombre, ponderacion, calificacion

    # Crear contenedor para los resultados
    output = widgets.Output()

    # Crear widgets para todas las tareas
    widgets_tareas = [crear_widgets_tarea(i) for i in range(1, 5)]

    # Botón para calcular
    calcular_btn = widgets.Button(description='Calcular Nota Final')

    def on_calcular_click(b):
        nonlocal total_ponderacion
        with output:
            output.clear_output()

            # Validar y recopilar datos
            tareas.clear()
            total_ponderacion = 0

            for i, (nombre, ponderacion, calificacion) in enumerate(widgets_tareas):
                # Validar nombre
                if not nombre.value.strip():
                    print(f"Error: El nombre de la tarea {i+1} no puede estar vacío")
                    return

                # Validar ponderación
                valido, resultado = validar_porcentaje(ponderacion.value, total_ponderacion)
                if not valido:
                    print(f"Error en Tarea {i+1}: {resultado}")
                    return
                total_ponderacion += resultado

                # Validar calificación
                valido, cal_resultado = validar_calificacion(calificacion.value)
                if not valido:
                    print(f"Error en Tarea {i+1}: {cal_resultado}")
                    return

                # Agregar al diccionario
                tareas[nombre.value] = {
                    'ponderacion': resultado,
                    'calificacion': cal_resultado
                }

            # Verificar que la suma sea 100%
            if total_ponderacion != 100:
                print(f"Error: La suma de las ponderaciones debe ser 100% (actual: {total_ponderacion}%)")
                return

            # Calcular y mostrar resultados
            nota_final = calcular_nota_final(tareas)
            print(generar_reporte(tareas, nota_final))

    calcular_btn.on_click(on_calcular_click)

    # Mostrar todos los widgets
    for widgets_tarea in widgets_tareas:
        for widget in widgets_tarea:
            display(widget)
    display(calcular_btn)
    display(output)

def version_consola():
    """
    Versión con entrada por consola
    """
    tareas = {}
    total_ponderacion = 0

    print("=== Calculadora de Nota Final ===")
    print("Por favor, ingrese los datos de las 4 tareas\n")

    # Solicitar datos de cada tarea
    for i in range(4):
        print(f"\nTarea {i+1}:")

        # Solicitar nombre
        while True:
            nombre = input("Nombre de la tarea: ").strip()
            if nombre:
                break
            print("Error: El nombre no puede estar vacío")

        # Solicitar ponderación
        while True:
            ponderacion = input("Ponderación (%): ")
            valido, resultado = validar_porcentaje(ponderacion, total_ponderacion)
            if valido:
                total_ponderacion += resultado
                break
            print(f"Error: {resultado}")

        # Solicitar calificación
        while True:
            calificacion = input("Calificación (0-5): ")
            valido, resultado = validar_calificacion(calificacion)
            if valido:
                break
            print(f"Error: {resultado}")

        # Agregar al diccionario
        tareas[nombre] = {
            'ponderacion': resultado,
            'calificacion': resultado
        }

    # Verificar que la suma sea 100%
    if total_ponderacion != 100:
        print(f"\nError: La suma de las ponderaciones debe ser 100% (actual: {total_ponderacion}%)")
        return

    # Calcular y mostrar resultados
    nota_final = calcular_nota_final(tareas)
    print(generar_reporte(tareas, nota_final))

# Detectar si estamos en Colab
try:
    import google.colab
    in_colab = True
except:
    in_colab = False

# Ejecutar la versión apropiada
if in_colab:
    print("Calculadora de la calificación final del curso")
    version_colab()
else:
    version_consola()
````
